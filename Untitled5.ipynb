```python
# Define flights (Flight ID, Start Time, End Time)
flights = [
    ('F1', 9, 11),
    ('F2', 10, 12),
    ('F3', 13, 15),
    ('F4', 11, 14),
    ('F5', 16, 18),
]

# Define crew members
crew_members = ['C1', 'C2', 'C3']

REST_TIME = 1   # minimum rest time between flights
```

```python
def is_valid_assignment(existing_flights, new_flight):
    """
    Checks NO OVERLAP + REST TIME constraint.
    existing_flights: list of tuples (id, start, end)
    new_flight: tuple (id, start, end)
    """
    _, ns, ne = new_flight

    for _, s, e in existing_flights:
        if not (ne + REST_TIME <= s or e + REST_TIME <= ns):
            return False

    return True
```

```python
#constraint checker function valid or not.
recursive_calls = 0

def backtrack_schedule(flights, crew, assignment, index=0):
    global recursive_calls
    recursive_calls += 1

    # If all flights assigned → success
    if index == len(flights):
        return True

    flight = flights[index]

    for member in crew:
        if is_valid_assignment(assignment[member], flight):
            assignment[member].append(flight)

            if backtrack_schedule(flights, crew, assignment, index + 1):
                return True

            assignment[member].pop()

    return False
```

```python
#crew-wise final flight assignment print .
assignment = {c: [] for c in crew_members}

start = time.time()
recursive_calls = 0

success = backtrack_schedule(flights, crew_members, assignment)

end = time.time()

if success:
    print("✔ Final Crew Assignment:\n")
    result = {c: [f[0] for f in assignment[c]] for c in assignment}
    print(result)
else:
    print("❌ No valid assignment found")

print("\nTotal Recursive Calls:", recursive_calls)
print("Execution Time:", round(end - start, 5), "seconds")
```

```python
def plot_gantt(assignment):
    fig, ax = plt.subplots(figsize=(10, 5))

    y = 0
    for crew, flights_list in assignment.items():
        for fid, s, e in flights_list:
            ax.barh(y, e - s, left=s)
            ax.text(s + 0.1, y + 0.1, fid)
        ax.text(0, y + 0.1, crew)
        y += 1

    ax.set_xlabel("Time")
    ax.set_ylabel("Crew Members")
    ax.set_title("Flight Schedule Gantt Chart")
    plt.show()


plot_gantt(assignment)
```

```python
import random

def generate_flights(n):
    flights = []
    time_slot = 8
    for i in range(n):
        start = time_slot
        end = start + random.choice([1, 2])
        flights.append((f"F{i+1}", start, end))
        time_slot += random.choice([1, 2])
    return flights


times = []
sizes = list(range(4, 11))

for n in sizes:
    test_flights = generate_flights(n)
    test_assignment = {c: [] for c in crew_members}

    start = time.time()
    recursive_calls = 0
    backtrack_schedule(test_flights, crew_members, test_assignment)
    end = time.time()

    times.append(end - start)

plt.figure(figsize=(8, 5))
plt.plot(sizes, times, marker="o")
plt.title("Execution Time vs Number of Flights")
plt.xlabel("Flights")
plt.ylabel("Time (seconds)")
plt.grid()
plt.show()
```

## NP-Hard Analysis

1. Airline crew scheduling is NP-hard.
2. Each flight can be assigned to any crew → exponential combinations.
3. Overlap & rest-time constraints prune the search tree.
4. Backtracking explores O(k × 2ⁿ) combinations.
5. Works only for small datasets.
6. Real airlines use heuristics or optimization solvers.
