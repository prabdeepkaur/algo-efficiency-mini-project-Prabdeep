```python
# -------------------------
# Capstone: Multi-Strategy Logistics Optimization
# Google Colab ready
# -------------------------

# Standard imports
import heapq
import itertools
import random
import time
import matplotlib.pyplot as plt

# -------------------------
# 1) Problem Modeling (Inputs & Constraints)
# -------------------------
# Nodes (warehouse + delivery locations)
nodes = ["Warehouse", "A", "B", "C", "D", "E"]

# Directed weighted graph (road network distances / times)
# If an edge missing -> treat as large cost (infeasible)
graph = {
    "Warehouse": {"A": 4, "B": 6, "C": 8},
    "A": {"B": 2, "C": 3, "D": 7},
    "B": {"C": 2, "D": 5, "E": 8},
    "C": {"D": 2, "E": 6},
    "D": {"E": 3},
    "E": {}
}

# Deliveries: (location, priority (1 high - 5 low), earliest_time, latest_time, demand)
deliveries = [
    ("A", 2, 9, 12, 1),
    ("B", 3, 10, 15, 1),
    ("C", 1, 8, 11, 1),
    ("D", 4, 13, 17, 2),
    ("E", 2, 14, 18, 1)
]

VEHICLE_CAPACITY = 3    # can carry 3 units
START_TIME = 8

# -------------------------
# 2) Helper: Dijkstra shortest paths from a source
# -------------------------
def dijkstra(graph, start):
    dist = {node: float('inf') for node in graph}
    dist[start] = 0
    pq = [(0, start)]
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, w in graph[u].items():
            nd = d + w
            if nd < dist[v]:
                dist[v] = nd
                heapq.heappush(pq, (nd, v))
    return dist

# Example - distances from Warehouse
# print(dijkstra(graph, "Warehouse"))

# -------------------------
# 3) Strategy 1: Greedy by priority + earliest deadline
#    (select deliveries up to capacity)
# -------------------------
def greedy_select(deliveries, capacity):
    # sort by (priority asc, latest_time asc)
    sorted_d = sorted(deliveries, key=lambda x: (x[1], x[3]))
    chosen = sorted_d[:capacity]
    return chosen

greedy_choice = greedy_select(deliveries, VEHICLE_CAPACITY)
print("Greedy selection (priority-first):", greedy_choice)

# -------------------------
# 4) Strategy 2: DP-like feasibility check for time windows (simple scan)
#    This is NOT full DP over subsets but a greedy feasibility using earliest times.
# -------------------------
def windows_feasible(start_time, sequence):
    # sequence: list of delivery tuples in execution order
    time = start_time
    for loc, priority, earliest, latest, demand in sequence:
        # travel time: from prev node (we assume starts at Warehouse each time for simplicity)
        # For more accurate model, sum pairwise shortest paths.
        # Here assume direct travel from warehouse for feasibility check (simplification).
        travel = dijkstra(graph, "Warehouse").get(loc, float('inf'))
        arrival = time + travel
        if arrival > latest:
            return False
        # wait until earliest if early
        time = max(arrival, earliest) + 0.5  # +service time 0.5 unit
    return True

print("Window feasible for greedy choice?", windows_feasible(START_TIME, greedy_choice))

# -------------------------
# 5) Strategy 3: Recursive route generation (permutations) + evaluate route cost
#    We generate permutations of the chosen delivery locations (or all if small n)
# -------------------------
def generate_routes(loc_list):
    # generate permutations of delivery nodes (excluding Warehouse)
    perms = list(itertools.permutations(loc_list))
    routes = []
    for p in perms:
        route = ["Warehouse"] + list(p) + ["Warehouse"]
        routes.append(route)
    return routes

def edge_cost(u, v):
    return graph.get(u, {}).get(v, float('inf'))

def route_total_cost(route):
    total = 0
    for i in range(len(route)-1):
        c = edge_cost(route[i], route[i+1])
        total += c
    return total

# use chosen nodes from greedy (location names)
chosen_locs = [d[0] for d in greedy_choice]
routes = generate_routes(chosen_locs)
print(f"Generated {len(routes)} possible routes for chosen deliveries.")

# evaluate routes and pick best valid (finite-cost)
best_route = None
best_cost = float('inf')
for r in routes:
    cost = route_total_cost(r)
    if cost < best_cost:
        best_cost = cost
        best_route = r

print("Best route (based on distance) for greedy set:", best_route, " cost:", best_cost)

# -------------------------
# 6) Brute-force full-search (small n) for comparison
#    Evaluate all subsets upto capacity and all permutations to pick globally best route under capacity
# -------------------------
def all_subsets_best_route(deliveries, capacity):
    best = None
    bestc = float('inf')
    n = len(deliveries)
    # iterate over all non-empty subsets
    for r in range(1, 1<<n):
        subset = []
        total_demand = 0
        for i in range(n):
            if (r>>i) & 1:
                subset.append(deliveries[i])
                total_demand += deliveries[i][4]
        if total_demand > capacity:
            continue
        locs = [d[0] for d in subset]
        # evaluate permutations
        for perm in itertools.permutations(locs):
            route = ["Warehouse"] + list(perm) + ["Warehouse"]
            cost = route_total_cost(route)
            if cost < bestc:
                bestc = cost
                best = (route, subset, cost)
    return best

start = time.time()
bf = all_subsets_best_route(deliveries, VEHICLE_CAPACITY)
end = time.time()
print("Brute-force best (subset+permutations) result:", bf)
print("Brute-force time:", round(end-start, 4), "seconds")

# -------------------------
# 7) Visualization: priorities bar and route costs
# -------------------------
# Priority bar
locs = [d[0] for d in deliveries]
pr = [d[1] for d in deliveries]
plt.figure(figsize=(6,3))
plt.bar(locs, pr)
plt.title("Delivery Priorities (smaller = higher priority)")
plt.xlabel("Location")
plt.ylabel("Priority")
plt.show()

# Route cost distribution (for chosen set)
costs = [route_total_cost(r) for r in routes]
plt.figure(figsize=(6,3))
plt.plot(range(len(costs)), costs, marker='o')
plt.title("Route costs for permutations (chosen deliveries)")
plt.xlabel("Permutation index")
plt.ylabel("Cost (distance)")
plt.show()

# -------------------------
# 8) Complexity comments (print summary)
# -------------------------
print("\n--- Complexity Summary ---")
print("Greedy selection: O(n log n) for sorting n deliveries.")
print("Dijkstra (single-source): O(E + V log V) using heap (repeated as needed).")
print("Generating permutations for k chosen deliveries: O(k!) time to evaluate all orders.")
print("Brute-force over subsets + permutations: O(2^n * n!) worst (infeasible for moderate n).")
print("Thus overall problem has combinatorial explosion â†’ relates to NP-hard problems like TSP/VRP.")

# -------------------------
# 9) Save/Reference the Capstone PDF (local path)
# -------------------------
CAPSTONE_PDF_PATH = "/mnt/data/Capstone Assignment.pdf"
print("\nCapstone assignment file available at (local path):", CAPSTONE_PDF_PATH)
# you can download or link this when preparing your GitHub repo / submission
```
